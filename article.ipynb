This article reports on a prototype digital laboratory for text analysis and discovery of interpretative pathways through historical materials related to Galileo Galilei (1563-1642). GaLiLeO: Galileo's Library and Letters Online prioritizes relationships not typically captured in database design for archival material. Drawing on best practices for humanistic modeling (<cite data-cite="1584927/PVFRKDSE"></cite>,<cite data-cite="1584927/4JAXCGB8"></cite>), these tools for exploration, question making, and question answering are inspired by a common epistemological practice in Galileo's lifetime: the juxtaposition of objects (often in a *wunderkammer* or cabinet of curiosities). New knowledge arises from seeing a document in comparison to others, not solely from the inherited value attached to an author, date, or organizational logic. Embedded functions allow a user to contextualize a document within a broader corpus of personal and semi-public letters, prefatory letters, and book-length texts from Galileo's library. By providing these contextual details, the digital space becomes experiential and representational, experimental and demonstrative. Accordingly, the preliminary text analysis tools in GaLiLeO represent a prototype to deprioritize post-hoc organizational structures in order to emphasize relationships across documents that may have resisted or otherwise been excluded from such structures.

## Introduction

This article engages with questions related to the usefulness of an unfinished (yet entirely functional) project that was designed to be experimental and iterative. Rather than navigate the tension between a computational toolkit and a digital text repository, Galileo's Library and Letters Online (GaLiLeO) operates from a dual premise: both that the documents must necessarily change with the expansion of digitized materials and also the hope that the tools can be applied to these newly configured corpora and corpora beyond the study of Galileo Galilei (1564-1643). GaLiLeO participates in three epistemological ecosystems: the current interdisciplinary field of Galileo scholarship, the 19th-century organization and curation of primary documents, and the 17th-century methods of knowledge making practiced by Galileo and his contemporaries. Importantly, this article will evaluate the extent to which humanistic design priorities can shape computational methods that foreground historical relationships across documents without remediating post-hoc interpretations of Galileo, his contemporaries, or his ideas.

Developed most intensively in 2016-2018, the GaLiLeO prototype and the process of its creation offered a way to brainstorm about a larger-scale project, work with a cohort of experts in various areas of 17th-century Italian culture, and draft new digital and computational analytical tools alongside scholarly editorial apparati. The resources are not limited by the confines of physical archives and offer both the flexibility and the challenges of decontextualizing digital representations of primary source documents with unique institutional and material histories. The goal was the contextualization, comparison, and expansion of points of entry for primary sources in early modern Italy, keeping in mind students new to the material and peer scholars in other fields looking for a way to investigate their own texts. In that regard, GaLiLeO addresses a point raised by Ryan Cordell during a visit to Bowdoin College in 2014: What happens when we do not know the person, term, or year that might be interesting in a corpus of texts? A further motivating question responds to evolving concerns in Digital Humanities and other fields: To what extent does digitized preconceived knowledge (complete with its accuracies and biases) prevent scholars from seeing interpretive pathways in a corpus? For the GaLiLeO designers, the result was a type of digital laboratory: a series of tools for exploration, question-making, and question answering. 

In spite of being able to publish new scholarship on Galileo as a result of these tools (<cite data-cite="1584927/TUDXHUQF"></cite>), as of August 2022, this project is on hold due to two primary obstacles. First, the underlying text from the National Edition of Galileo’s works reflects such significant editorial interventions in orthography, punctuation, and sometimes word choice that it is not a reliable marker of writing style. Second, the developers are searching for an outlet to publish the metadata associated with the correspondence, the full texts of materials from the library, and the code for running analysis. At the time of the most intensive project development, our local institutions were not in a place to support long-term scalable hosting of the project. While not necessarily an obstacle, the code ought to be optimized for runtime efficiency, since I am a Digital Humanist who codes, but not a software engineer. GaLiLeO is thus a project in search of scholarly and institutional collaborators for building a user-friendly interface, a robust underlying data set, and hosting. This article documents the project as it was demonstrated in September 2018, with slight modifications to function using the tools of this journal and with identification of improvements that could be made in the future.

## Project History

The idea for this project began as a conversation in April 2016 at Stanford University about the points of connection and similar goals of two collaborative efforts that had already been underway for several years: developing an editorial apparatus for Galileo's correspondence based on close reading of the available letters and creating a digital and computational library of the books that Galileo owned. From the perspective of History of Science, Hannah Marcus had worked with a team of scholars to identify thematic overlap in the surviving correspondence related to Galileo. Part of that project also involved careful attention to metadata about the letters, which has resulted in new findings. (<cite data-cite="1584927/5R7T42AP"></cite>) From the perspective of Digital Humanities, I have been working with colleagues at my institution to develop an analytical framework for digital artifacts that evaluates the role of the digital (representations in binary) and the computational (the practices enabled through digital representation). (<cite data-cite="1584927/LIZJVZTN"></cite>) We articulate our approach as a two-way street, in which those artifacts, for example a computational text analysis tool, can be applied to questions and objects of study in the Liberal Arts (Arts, Humanities, Social Science, Physical Sciences). But the tool can also be designed or critiqued with the priorities of those disciplines at the forefront. In this context, my practice of Digital Humanities is concerned with more than the application of computational tools in the study of humanistic expression. I also investigate the ways in which humanistic priorities can guide the development of such tools. With this in mind, Hannah and I explored the extent to which humanistic intervention could reshape the process of digitization, the construction of any underlying data structures for GaLiLeO, the design of the user interface, and the nature of the tools. This article will focus on the data and tools. Given our overlapping goals of learning more about Galileo and his contemporaries through the study of text, we began to design what became GaLiLeO: Galileo's Library and Letters Online.

Naming credit for GaLiLeO goes to Hannah Marcus and Morgan MacLeod after much discussion about what this project would be. We struggled with the identity of the hybrid tools-text nature of the platform. We were not trying to create a digital archive, which would emphasize findability and display of a range of related material. We could aspire to the William Blake Archive, but we were a small group without funding or broad technical support. We knew that we were not aiming for a digital edition, even though Marcus' team had created a rigorous and extensive scholarly interpretive layer for the correspondence. Using the vague "online" descriptor avoided these problems of identity. We hoped that it would gesture toward actions of aggregation, exploration, and experimentation that are enabled by the web. While bringing together plain, full text of the letters and limited contents from the library would create bridges across otherwise siloed resources, our aim was not to make a claim about a collection but to evaluate materials, methods, and outcomes of analyses of those texts.

Our enthusiasm quickly outpaced resources of time and funding. We composed an ambitious application for a Digital Humanities Advancement Grant through the National Endowment for the Humanities in the United States. (<cite data-cite="1584927/B5Q585LU"></cite>) The feedback on the failed proposal helped us to scale back our initial tasks to a proof-of-concept for the combined text analysis of the two corpora and their metadata. We maintained the same intellectual priorities outlined in the grant, but worked with materials at hand to focus more on method.

While current use of GaLiLeO is limited to the two principle investigators (and the exploration facilitated by mybinder functionality with a sample of the data for this article), it has always been intended for a broader audience of Galileo scholars, interested peers in history and literature, and students. Thanks to financial support from Harvard University's Dean's Competitive Fund for Promising Scholarship, we were able to invite scholars to campus to test this alpha prototype and brainstorm next steps. The resulting September 2018 workshop was an opportunity to evaluate our ground-up approach to humanistic intervention in digitization and computational analysis.

## The Digital Galileo Ecosystem

The largest digital resource for Galileo specialists is the galileo//thek@ (Galileoteca) developed and maintained by the Galileo Museum in Florence, Italy. (<cite data-cite="1584927/K7RFDQH5"></cite>) The Galileoteca provides access to primary documents, scholarly editions, and secondary scholarship. Like most of Galileo scholarship, the primary documents in the Galileoteca are heavily influenced by the late-19th century National Edition of Galileo's works and correspondence compiled by Antonio Favaro (1847-1922) and his team of collaborators. (<cite data-cite="1584927/XYL4T68J"></cite> )Documents are organized according to their location in the national edition and digitized text reflects Favaro's editorial interventions, most often in materials written by Galileo, rarely in texts authored by others. The excellent curatorial and editorial team at the Museo Galileo has added other interpretive layers to the digital interface to help visitors find the appropriate documents easily.

The other key resource in the digital Galileo ecosystem provides historical context for key people and ideas: The Galileo Project at Rice University. (<cite data-cite="1584927/JEPMCMF5"></cite>) Conversations with the specialists at the workshop confirmed that this early Digital Humanities project remains a frequent reference while we conduct our research. Among many other resources on the site, the chronological and bibliographic narratives provided by the project contributors bring to life key events and individuals that are frequent authors or frequently mentioned in the primary source documents.

GaLiLeO seeks to go below these *post hoc* layers and return to the texts to explore new interpretive pathways that have not been predetermined by centuries of scholarship. In the Galileoteca, specialists who know what document they want to find, those who are studying a chronological period in Galileo's lifetime, or have a thematic or topical research question in mind can use the resources to access related materials. Users can search word forms and lemmas within certain areas of this digital archive or across elevent kinds of materials that also include biographies, indices, and iconography. It is an excellent finding aid for those who already know the specific text they seek. As a complement to the Galileoteca, the tools presented in GaLiLeO push more toward emphasizing the choices humanists make based on contextual relationships and those same tools allow for the circumstances (the contextual elements) to intervene.

The GaLiLeO approach recognizes the ways in which digital infrastructures instantiate interpretive arguments. Literary and historical inquiry relies on context, often multiple contexts, which is challenging to evaluate in a large corpus if the underlying navigation principle is to quickly demonstrate single documents that best match queries. Since Hannah and I were bringing together two corpora that are not typically studied together, our interpretive argument was that the broader print culture surrounding Galileo and his correspondents influenced the letters and likely vice versa. This meant creating textual models and query possibilities that would highlight similarities or differences in a way that gives researchers enough context to both evaluate influence and follow leads suggested by those circumstances.

The availability of data was one of the limiting factors in further development of the project. Even for an individual as famous as Galileo (or perhaps because of it), there is no freely-available set of plain-text files for analysis. The correspondence in the National Edition of Galileo’s works included over 4,300 letters from the 20 volumes edited by Favaro in the 19th century (<cite data-cite="1584927/XYL4T68J"></cite>). Many letters were only excerpted in Favaro's edition. While a scan of the first printing of the National Edition is available, images of manuscripts are housed by different institutions with strict reproduction rules. Material details of papers and books are sparse, as is the information about the development of the specific archives over time. One participant in our workshop spoke about the ideal humanistic digital laboratory offering access to a searchable corpus, allowing for a choice of transcriptions, and images of the originals with an emphasis on physical properties. While many such digital resources exist for an Italian figure like Dante, this goal is still elusive for Galileo studies. 

Moreover, the field of Galileo studies is moving in directions that still require the attentive philological study of what Galileo and his contemporaries wrote, but with the integration of other disciplinary perspectives. Some of the new schoarly work on Galileo focuses on visual culture, both in terms of diagrams and illustrations but also the arts. Renewed attention has been brought to Galileo's marginalia and reading practices. Researchers are also pushing back against claims of exceptionalism to highlight linguistic traditions in which Galileo participated and also the popular topics that he avoided. Ultimately, embedding the project within the digital Galileo ecosystem would benefit the field more than creating a new digital history tool with claims to being something definitive that incorporates all materials. Using this tool in conjunction with the lemmatized resources in the Galileo//thek@ or the names in The Galileo Project would still maintain the experimental, exploratory spirit of the GaLiLeO tools. The Archilet - Reti Epistolari project has already created a model for this with metadata related to literary correspondence in Italy 1600-1800. (<cite data-cite="1584927/EPV67BFC"></cite>) While GaLiLeO cannot yet address these aspects of embeddedness, we feel that the motivated user will use the output from our tools as search queries in this larger ecosystem.

## GaLiLeO and Hermeneutica

One might ask why we did not repurpose a text or metadata analysis tool already in existence. After all, Voyant Tools was one of the primary inspirations for the functions that I wrote for GaLiLeO. (<cite data-cite="1584927/2WCMWWUK"></cite>) The theoretical framing for Voyant, provided by Rockwell and Sinclair in *Hermeneutica*, is structurally integral to how we teach text analysis at Bowdoin. (<cite data-cite="1584927/PVFRKDSE"></cite>) In DCS we talk about Abstractions, with an emphasis on models necessarily built into digital artifacts. Even prior to this is the recognition that the digitization process already represents an abstraction from the reality of the physical letter or book to its representation as code. In the best of cases, the digital and computational artifacts, which are the tools of analysis, address more than just an abstraction of a computational process, but also the other humanistic models that Rockwell and Sinclair identify in their exploration of three models in digital humanities: "the theoretical model of the problem or cultural phenomenon; a working model of an instance of the phenomenon based on the first; and a model of the process whereby we will get the second." ((<cite data-cite="1584927/PVFRKDSE"></cite>), 163) For literary and historical work this third model often requires the analysis of the interplay of content and metadata, contextualizing expressions against a backdrop of other corpora, and considering the role of materiality in the interpretation of a text. The existing tools necessarily draw our attention to certain features while deprioritizing others. 

For GaLiLeO, both our analytical goals and the size of our corpus (thousands of documents), made repurposing Voyant challenging. The contextual tools in Voyant draw attention to high-frequency terms moreso than unusual ones. The "Distinctive Words" feature in the "Summary" pane does higlight (literally and figuratively) ways in which authors depart from one another in lexical choices.

Voyant's "Distinctive words" list in the figure above helpfully shows how document EdNaz574, for example, differs from the others that use the word *ochiale* (an alternative spelling for the Italian word for spyglass or telescope). I will return to the corpus analyzed in Figure 1 in the evaluation of the text and lexical tools in GaLiLeO, all of which use this variant of the more standard *occhiale*. For now, I will just point out the document EdNaz574, a letter written in 1610 that will be the example carried throughout this analysis. The GaLiLeO tools brought this unusual orthography to light, rather than obscuring it behind editorial practices and database elements of lemmatization and orthographical standardization. Lemmatization and certain editorial practices collapse these variants into one form for searching for all of the letters that talk about the telescope such that a search for *ochiale* produces no results in the Galileoteca. With a term like the related *telescopio*, which has several hundred results, even the list needs to be parsed to make sense of the kinds of use of the word. Computation can help.

Consequently, rather than a model of the documents related to Galileo, the tools presented here provide a model for asking questions in humanistic fashion. The approach borrows this phrasing from Willard McCarty, who distinguishes between an archeological map *of* a historic site (i.e. the reconstruction of the National Edition of Galileo's works) and the blueprint *for* a new edifice (GaLiLeO). (<cite data-cite="1584927/4JAXCGB8"></cite>) A model *of* Galilean texts would be forever incomplete and many parts already exist in the current digital ecosystem. A model *for* inquiry on the other hand would be extensible to the new data sets, for Galileo, but potentially for other authors and collections. The tools in GaLiLeO thus embrace the epistemology of *Hermeneutica*, but not the implementation. Since, as Rockwell and Sinclair point out, an "instrument implements a theory of interpreting the phenomenon it was designed to bring into view" (<cite data-cite="1584927/PVFRKDSE"></cite>, 162), we needed a new instrument. Historians are seeking mechanisms to understand tone, reliability (through consistency), corroboration with other viewpoints, and verification. Literary scholars engage with the multiple contexts through which an expression conveys potentially different meanings. For this we needed to bring together not just words and documents, but the overall corpus attributes. Again, as Rockwell and Sinclair so efficiently state: "developing tools involves bearing hermeneutical theories." (<cite data-cite="1584927/PVFRKDSE"></cite>, 161) 

This meant that the organization of the texts had to co-evolve with the analytical needs based on the quality and groupings of texts and their associated metadata. The proof of concept identified several areas for future development in terms of customization: how to interact with the specialist editorial layer (or not), choosing a subcorpus of interest, limiting or expanding the number of languages considered, and how to handle Galileo's work. As presented at the workshop, the overall corpus could be subset into specialized document groupings based on type of document and metadata tag categories. The prototype was built in a Jupyter notebook, which tried to offer novice coders instructions for use, but the project would benefit from a more user-friendly interface.

As Rockwell and Sinclair point out, these hermeneutica draw our attention to "the materiality of interpretation." (<cite data-cite="1584927/PVFRKDSE"></cite>, 152) In Digital and Computational Studies, our analytical framework similarly draws attention to what actions the digital artifact allows and what actions are omitted that could have been there. For example, is a tool designed for scrolling (accumulation) or for simultaneity (comparison and juxtaposition)? For GaLiLeO the materialities rest on two kinds of instantiations of interpretation via juxtaposition of text against others: providing context around specific documents or words and experimenting with pattern-based inquiry. The contextual output draws attention to the marginal, mainstream, and missing information related to a query. Data about commonalities and differences appear together by default. Pattern-based functions, currently at the end of the prototype, push the materiality of interpretation further by asking the scholar to frame a question without knowing a specific document, word, or author that might meet the criteria.

## GaLiLeO's Tools and the Experience of Strangeness

This approach comes with two caveats. The first results from Italian Jesuit Roberto Busa's advocacy for modeling the "conceptual system" of a text's author and the context of that author. (Paraphrased in <cite data-cite="1584927/PVFRKDSE"></cite>, 152-153). It is a reminder to avoid building tools that apply our own, positivist (i.e. 19th-century), modernist, or post-modernist viewpoints to the detriment of seeing and hearing the system of connections between texts as they were created and experienced in their own historical moment. In other words, I strove to design tools that embraced Italian cultural theorist and historian Carlo Ginzburg's theory of "Making things Strange." (<cite data-cite="1584927/YJWVTQER"></cite>) Seeing the text differently by decomposing it and visualizing its contents symbolically creates the cognitive distance for rupturing potentially limiting views on our texts. Stephen Ramsay has argued that this kind of algorithmic tranformation "can provide alternative visions" of a text that give rise to deeper readings. (<cite data-cite="1584927/UTT8CW8L"></cite>, 16) Rather than a remediation of post-hoc lenses with their own epistemological priorities, I want this tool to identify for me in a way that I, as a 21st-century scholar cannot, the categories, organizing principles, affinities, disparities, patterns, and outliers in the objects that I have collected.

The second caveat builds on this principle and comes from Galileo and general practices of thinking through objects in the Seventeenth Century (<cite data-cite="1584927/YUYR8QY3"></cite>). This also echoes the concerns raised in *Hermeneutica*. In response to one of Stanley Fish's notable critiques of DH, Rockwell and Sinclair respond in a way that echoes Galileo's pushback against opponents of his new science. The Voyant creators write that Fish's argument is "like trying to understand railroads by reading books about them instead of looking at models or taking a ride on a Shinkansen" (<cite data-cite="1584927/PVFRKDSE"></cite>, 151, referencing the Japanese bullet train). Galileo would agree. He offers a more elaborate explanation in the voice of one of the speakers early on the First Day of the *Dialogues on the Two Chief World Systems* (1632): "there are those who possess all of the precepts of da Vinci, but would not know how to paint a stool. Playing an organ is not learned from those who know how to make organs, but from those who know how to play them... painting is learned from continuous drawing and painting." (<cite data-cite="1584927/XYL4T68J"></cite>; my translation from VII, 69-70) He goes on, but the point is the same - reading about an instrument and knowing how to use it are two different things. In the case of an archive informed by previous scholarship, this inherited knowledge is valuable; being able to access the philogical and analytical discoveries of previous scholars through such a database structure is indeed a rich resource. My argument is that we can do more. Galileo's suggestion that to be a philosopher, or for us to be humanists more generally, we need to act and practice our skills, implies the need for a tool that augments our actions. To continue Galileo's metaphor, in order for our work as humanists to be more than great practice over previous humanists' work, reified in digital form, rather than build the database (the organ in the analogy), we must continually practice our skills of discovery, uncovering, recreating, differentiating, evaluate, contextualizing, and documenting representations and expressions. 

Galileo's lesson about knowing by doing rather than knowing by reading about was deeply specific to the controversies in which he was embroiled over inherited knowledge, particularly that of the classical Greek philosopher Aristotle. His resistance to following the minds of others in philosophy came from experience and experimentation with new tools such as the telescope. A primary mechanism employed by Galileo in many of his experiments, and adopted in the tools of Galileo was juxtaposition. (<cite data-cite="1584927/GT5XL8MF"></cite>) Early modern scholar and digital humanist Brent Nelson is exploring the digitization of this new empiricism with objects in cabinets of curiosities (<cite data-cite="1584927/4G55RQP2"></cite>). GaLiLeO attempts to do this with text. Moreover, the ethos of skepticism, also expressed in *Hermeneutica*, can also address 21st-century Digital Humanities priorities of de-colonizing archives. In order to know what is interesting enough to search in a corpus, a user needs to arrive with prior knowledge of the values associated with its contents. While not always problematic, systemic exclusion, majority perspectives, and other kinds of bias limit the interpretive pathways of a collection of text structured by technology.  

In order to provoke the sensations of strangeness and provide an alternative entry point to the corpus of Galileian materials, we wanted to be able to ask contextualized questions through computation as much or more than we wanted to retrieve specific information that had been previously identified as somehow more important than the rest. This guided the data creation process. Andrew Piper has recently articulated this preoccupation with presumed moral superiority of inherited truths as the "curse of Lorenzo Valla." (<cite data-cite="1584927/942BYLHC"></cite>, 10-13) Preliminary, contextual analysis in GaLiLeO occurs prior to the user arriving at the interface based on lexical, quantitative information without ontological restrictions or siloes of form, genre, or period that would have been alien to an early modern reader. (This also solves a runtime problem, similarly to the functionality of the Text Analysis and Concording Tool (TACT), but of course restricts usability by this predefinition. (<cite data-cite="1584927/PVFRKDSE"></cite>, 57)) The code is not elegant, but captures quantitative information about each document, the words in the document, and the author. As shown in the figure below, the underlying metadata is also contextualized within the corpus. The results are stored as RDS objects to be loaded into memory when the user chooses a subcorpus for exploration.

Figure 3 outlines the specific and relative information created and saved for each document. Again, there are redundancies and inefficiencies since this is an alpha-prototype coded by a humanist. Many of the attributes are recorded with variations: tokenization includes and excludes punctuation, stylistic similarity based three different feature classes and two distance measures, as well as distinguishing between documents written by Galileo and those written by other authors. The goals were twofold: achieve as many perspectives as possible on the texts and word types as well as explore the strengths or limitations of each perspective in comparison to the others. The method allows for humanistic experience of the texts as well as points to what Ted Underwood has called perspectival modeling and I have referred to as computational parallax, again building on 17th-century knowledge making practices. (<cite data-cite="1584927/4HDB2DPJ"></cite>, <cite data-cite="1584927/WKJA6LJE"></cite>) Creating many of the features also allows for querying by attribute range rather than the specific attribute. For example, by saving the contextual information, a user can identify authors who wrote many documents in the corpus, but were the recipient of none.

The end goal is not definition or specification, but experience. GaLiLeO aspires to providing pathways of interest, not interpretations. Much like the ARTFL Encyclopédie (<cite data-cite="1584927/DMRYV5NK"></cite>), that uses multiple digital models for users to explore, a preliminary topic model was made from each corpus for proof-of-concept purposes. (<cite data-cite="1584927/AKKT9WCI"></cite>) Our goal for the workshop was to see if we could build even an elementary method for prioritizing the kind of question-making that would open new doors onto the texts and their authors.

Typical database design would require the eventual user to have one of two things:
- prior knowledge of key actors in the period, important dates, sites of intellectual exchange, and significant keywords (all likely brought to light by previous scholarship).
- enough curiosity to provoke browsing ordered lists of people, dates, places, or concordances in the hopes of finding something interesting.

While our underlying data does capture information about people and places (with Named Entity Recognition to be implemented), it also tries to bring to light the bigger picture in which they appeared. By adopting an experiential approach with a perspective of strangness, GaLiLeO gives relevance to inconsistencies, prizes the anomalous, renews the priority of the individual text or author, draws attention to gaps, and capitalizes on juxtaposition to provoke new inquiry.

Creating an experiential, if not strange, opportunity from a list of author names would require a more ambitious model of the subcorpus than was available to us. For example, it would be helpful to both an expert scholar and a student to be able to search by the following attributes that can be calculated for each named person or place:

- that a letter recipient is only present in paratext in the library
- if a person mentioned has a higher centrality in the library than in correspondence
- that the destination of the letter is the most frequent city to which letters were sent; the second-most frequent city in which books in the library were published

By creating this metadata for documents, we facilitate structuring a query that identifies relationship intensities (central, peripheral or eccentric), frequencies (maximum, high, low, minimum, none), and document categories (private, public, print tradition in the library) of interest. In this sense, we respond to a challenge that Ryan Cordell identified in the development of the large-scale digital text archives: we often do not know what to put in a search bar. (<cite data-cite="1584927/VNEYQCPL"></cite>)

Scholars and students do know what kinds of questions are interesting, and the GaLiLeO team hopes to have built something to model that kind of query. In that sense, GaLiLeO becomes a site of discovery, not simply transmission. The hierarchies often embedded in digital tools through remediation of archives are brought onto the same plane via the underlying data structure. By changing identifiers related to documents (not alphabetical, chronological, or numerical), we place them in an interconnected space of representation and understanding (thus they are actual relations rather than a relational table with connections assigned through inherited knowledge). The user can experience these texts with a perspective that is not her own - a browsable structure that provokes inquiry, allows for the selection of a virtual book from this network and reveals all of the ways that it nests within hierarchies or stands apart from them. Rather than browse by a post-hoc genre category, the adjacencies of texts based on a variety of features can be evaluated for their metaphorical rough edges, apparent lack of connection, the ambiguity of the language that binds them, and the serendipitous viewpoint that their proximity allows.

Admittedly, the distinction between counting punctuation or not might be negligible. However, Galileo's prose is built on clauses and subclauses, so counting words followed by punctuation as a token helps us to capture something about the prominence and variation within that stylistic element. For example, a document 1000 words long with 500 unique word types has a TTR of 50%. If the letter writer uses punctuation to establish rhythm and structure in her sentences, at a rate of one mark for every 5 words, the number of unique types could bump up to as high as 700, while the length remains the same. Then the TTR would be 70%. If she has long sentences structured around connective words rather than punctuation, the number of types will only increase by a small percentage. Such measurements invite more opportunities to design pattern-based inquiry options.

In contrast, letter writers who use Galileo's preferred *occhiale* (not pictured) over time seem to consistently have done or made something (*fatto*), rationalize when they speak about it (*perchè* - because), and discuss a gain or excess (*molto* and *più*). 

The importance of a word is characterized by these functions as more than a highest frequency value or as a marker of distinction (such as tf-idf). By contextualizing the term within textual and metadata information, a user can ask questions about patterns without knowing the words that fit those patterns. The visual and quantitative display create a critical distance from the text that can provide an alternative perspective on the lexical relationships embedded in the text. This invites surprise, questioning, and potentially evaluating previous claims of importance in a collection. The final tools in GaLiLeO attempt to make this explicit, but are ultimately quite simplistic in the prototype.

## Pattern Tools in GaLiLeO

Where the document and lexical tools can help to identify a potentially fruitful avenue of inquiry, they are aided significantly by some contextual knowledge on the part of the user. Alternatively, the more ambitious pattern-based tools attempt to foreground the kinds of questions and structures that can motivate historical and literary analyses. These final tools still need significant development and refinement.  

One function allows direct comparison of authors' vocabulary. The eventual goal is to better understand lexical variety, breadth, and overlap. The first step, which was intended to provoke conversation at the prototype workshop, simply reports on the size of the vocabulary in the texts in the subcorpus by a given author. To be more helpful, it should also output the number of documents and the length of those documents. In the example below, the function is called three times to compare Galileo's lexicon to that of his daughter (Suor Maria Celeste) and Margherita Sarrocchi.

```r
compare_author_vocabulary("Galilei, Galileo", corpus, text_attributes, corpus_attributes)
```

```r
compare_author_vocabulary("Galilei, Maria Celeste", corpus, text_attributes, corpus_attributes)
```

```r
compare_author_vocabulary("Sarrocchi, Margherita", corpus, text_attributes, corpus_attributes) 
#Replace the Xxxx with Last name, First name, please leave the ""
```

In this subcorpus Galileo has 5 times as many letters as his daughter. Maria Celeste's 64 is already 16 times greater than Sarrocchi's. Their vocabularies reflect slightly different proportions: Maria Celeste's is nearly one quarter of Galileo's; Sarrocchi's is one seventh of hers. As with nearly every other feature, the focus is on contextualizing Galileo against the backdrop of other authors in the corpus or subcorpus. A version of this function is already provided in the document-level information, where we learned that Sarrocchi used an alternate orthography for the Italian terms for academics and mathematics in letter 574 in the National Edition of Galileo's work. This function expands those results to all documents by the author and, at least at this preliminary stage, allows for comparison by range of use, not simply single occurrences in a document.

The final tool is perhaps the most exciting. What are the words that Galileo uses in high frequency that the other authors in the corpus do not use at all? The function “find_types_by_range” provides that output. Again, this is a proof of concept, which needs more customization. For now, high is considered to be the 75th percentile or higher in frequency of usage; low is the 25th percentile. Users can also input a simple yes or no to indicate words that Galileo or other authors use (yes) or not (no). 

```r
#Takes a few seconds to run
find_types_by_range(corpus_attributes, GG="high", NotGG = "no", size =25)
```

Rather than making the familiar words of the corpus strange, the tool simply provoked a sense of the magnitude of the unexplored. The results are cumbersome, given the state of the underlying textual data. Similarly Patrick Juola and Ashley Bernola's Conjecturator (<cite data-cite="1584927/AJ5ZCGGE"></cite>) that compares textual features such as the use of a word within two groups of texts that are differentiated by a structural aspect such as decade. As with the over 87,000 statistically significant observations, these results are "mind-numbing, even demoralizing." (<cite data-cite="1584927/PVFRKDSE"></cite>, 135) As instantiated, the output still requires specialist knowledge to select a potentially useful path forward. Further interactivity and use of visual cues would help to address some of these concerns.

With a tool such as this an author would not need to fit or fight a model of a collection of texts; she would have her own. By seeing these permutations of terms, we can start to ask questions about individual models of language use rather than a best-fit model for all authors represented. Sarrocchi's document can be compared to the authors' corpus and the corpus at large. The topic model of the overall corpus would be deprioritized in order to see and compare individual authors and texts. In this way, we might capture some of Underwood's perspectival modeling without risking the bias identified by Katherine Bode (<cite data-cite="1584927/5RCFENA7"></cite>). Perpetual reference and comparison to a model that is heavily influenced by the prolific authors found within it continues to disenfranchise the already marginalized figures who were nevertheless writing and creating in the period. If we continue with the example of Sarrocchi, we can imagine a future instance of GaLiLeO in which we could compare her letters and the text of her epic poem (with which Galileo was familiar) to Galileo's corpus or to the library corpus overall. Her Latin geometrical and theological texts (should they be found) could be distinguished from her letters and text in Italian for a more nuanced comparison of stylistic, thematic, lexical, and contextual similarities or differences.

## Implications and Next Steps

The day-long workshop generated exciting ideas about what a project like this could be. The energy and interest were palpable. We learned about a diverse range of digital user needs that went far beyond the scope of GaLiLeO and we identified immediate and persistent obstacles. We realized immediately at the workshop, and later relived when coauthoring our article about Galileo's use of *cristallo*, that the ability to see the words contextualized at multiple levels of address both identified points where a collaboration was necessary and how the tools facilitated that collaboration. For this literary scholar and digital humanist, I could see the poetic and dramatic points of contact with the texts and narratives that were most familiar to my historian of science collaborator for other attributes. With someone like Sarrocchi, this reveals that the historian might consult with an art historian to better understand the similarity with the painter Cigoli, a social historian or historian of science specializing in the creation and reception of institutions for validating knowledge in early modern Italy, and a literary scholar for the connections between her letters to Galileo and the types of expression found in her epic poem.

The future of the project is still uncertain, pending time and also funding. Here are the general pieces of feedback that we received:
- More tools would be welcome: an API, historical text reuse detection, reporting on avoided topics, named entities, and a “humanistic suggestion tool” that offers models for queries along with rationales.
- More documents would be welcome: marginalia, more library texts, better OCR, and machine readable text based on manuscripts instead of 19th century editorial versions of letters.
- Transmedia linguistic analysis would be a valuable further step.
- GaLiLeO represents an opportunity to bring new perspective on digital projects that are considered “complete” by linking from its terms, people, years, and texts to the lemmario, biographies, chronologies, and bibliographies that have been developed by colleagues in Italy and the U.S. These include the Galileo//thek@, The Galileo Project, and even the authority file of 8,000 names connected to Galilean materials created by the Museo Galileo (http://www.museogalileo.it/esplora/biblioteche/biblioteca/archivionomi.html){:target="_blank"}
- The project will benefit from phased roll-outs of the underlying data and metadata, particularly as more texts from the library are added and additional metadata tags are created.
- This is an opportunity to involve undergraduate and graduate students of Italian in an exploration of the history of Italy as well as the ways in which Italian content can influence the development of predominantly Anglophone technology. It could also be a platform to put specialist, informed readers in contact with novice, curious readers.

Beyond this, the plan for GaLiLeO anticipates that any word in the letters or library is:
- a string of alphanumeric characters (ngram)
- a potential editorial tag provided by specialists
- a part of speech (POS)
- a keyword in multiple textual contexts
- a piece of a textual structure (sentence, verse, speaking part, marginalia)
- a piece of multiple semantic networks
- an abstraction (indicative of a dictionary of synonyms and related terms)
- a concept (a collection of instances across time and contexts)
- an expression in time
- a position in vectors of words
- part of metadata
- part of text or paratext
- a material object with a print and/or manuscript history and visual qualities
- an instance of a frequency of usage:
    - In the corpus
    - In works written by Galileo
    - In works written by others

As such, I continue to build tools that allow visual or numerical points of entry for understanding words in their multiple significances in the collection of documents. The GaLiLeO team imagined a dashboard of widgets similar to Voyant, ultimately more user-friendly than the Jupyter notebooks we used to demonstrate the prototype to our colleagues. Among other new widgets to develop, I would like to find a better way to visualize co-occuring words over time, likely through incorporation of a diverging bar plot.

The computational and quantitative results become a window onto the collection and indication of fruitful pathways, rather than a declaration that foreclose interpretation. By providing a way to enter the corpus without the 19th-century interpretive lens, scholars should be able to benefit nonetheless from that philological and editorial work along previously undefined paths of inquiry.
